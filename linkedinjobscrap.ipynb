```python
import selenium
from selenium import webdriver
import time
import pandas as pd
import os

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
```

```python
url1='https://www.linkedin.com/jobs/search/?currentJobId=3980897289&geoId=114806696&keywords=Marketing%20Data%20Analyst&origin=JOBS_HOME_LOCATION_AUTOCOMPLETE&refresh=true'
```

```python
service=Service(executable_path=r'C:\Users\sony\Desktop\fullstack-code\chromedriver-win32\chromedriver.exe')
options=webdriver.ChromeOptions()
driver=webdriver.Chrome(service=service, options=options)
driver.get(url1)
driver.implicitly_wait(10)
```

```python
n=driver.find_elements(By.CLASS_NAME,'results-context-header__job-count')[0].text
```

```python
n
```

```python
y=pd.to_numeric(n)
y
```

```python
i=2
while i<=16:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
    i=i+1
    
    try:
        x=driver.find_elements(By.XPATH, "//button[@aria-label='Load more results']")
        driver.execute_script("arguments[0].click();",x)
        time.sleep(3)
    except:
        pass
    time.sleep(4)
```

```python
companyname=[]
```

```python
for j in range(y):
    try:
        company=driver.find_elements(By.CLASS_NAME, 'base-search-card__subtitle')[j].text
        companyname.append(company)
    except IndexError:
        print("Done")
        
```

```python
companyname
```

```python
titlename=[]
```

```python
for j in range(y):
    try:
        title=driver.find_elements(By.CLASS_NAME, 'base-search-card__title')[j].text
        titlename.append(title)
    except IndexError:
        print("Done")
```

```python
titlename
```

```python
companyfinal=pd.DataFrame(companyname,columns=["Company"])
```

```python
titlefinal=pd.DataFrame(titlename,columns=["Title"])
```

```python
final=companyfinal.join(titlefinal)
```

```python
final
```

```python
final.to_csv('linkedinJobDetails.csv')
```

```python
linkList=[]
findLink=driver.find_elements(By.CLASS_NAME,'base-card__full-link')

for k in findLink:
    linkList.append(k.get_attribute('href'))
```

```python
linkList
```

```python
linkfinal=pd.DataFrame(linkList,columns=["Link"])
```

```python
final1=final.join(linkfinal)
```

```python
final1
```
